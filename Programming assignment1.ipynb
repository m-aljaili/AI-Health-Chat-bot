```python
import math
a = 0
b = 0
step = 0.01
c_threshold = 0.01
```

```python
x=[1,2,4,6,8]
y=[2,5,6,9,11]
```

```python
# mse = sigma (y - (ax + b))^2
# mse derivative by a => sigma 2 *(y - ax - b) * (-x) should be  0
# mse derivative by b => sigma 2 *(y - ax - b) * (-1) should be  0

def mse(a, b, x_set, y_set):
    diff = 0
    
    for x, y in zip(x_set, y_set):
        y_ = a*x +b
        diff += (y - y_)**2
        
    return diff/len(x_set)

def gda_a(a, b, x_set, y_set):
    gradient_a = 0
    
    for x, y in zip(x_set, y_set):
        gradient_a += 2*(y - a *x -b) * (-x)
        
    return gradient_a / len(x_set)
        
def gda_b(a, b, x_set, y_set):
    gradient_b = 0
    
    for x, y in zip(x_set, y_set):
        gradient_b += 2*(y - a *x -b) * (-1)
        
    return gradient_b / len(x_set)
```

```python
idx = 0
while(1):
    idx += 1
    loss = mse(a,b,x,y)
    
    gradient_a = gda_a(a,b,x,y)
    gradient_b = gda_b(a,b, x, y)
    
    a = a - step * gradient_a
    b = b - step * gradient_b
    
    e= math.sqrt(gradient_a**2 + gradient_b**2)
    
    print("{0} loss: {1} a: {2} b: {3}".format(idx, loss, a, b))
    if(e < c_threshold):
        break
```
